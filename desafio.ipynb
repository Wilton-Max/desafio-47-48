TABELA DE INVESTIMENTOS EM TV, RADIO E JORNAIS

```python
import pandas as pd
import matplotlib.pyplot as plt
tabela = pd.read_csv("advertising.csv", sep=",")
print(tabela)
```

TOTAL INVESTIDO NA TV:

```python
total_tv = tabela["TV"].sum()
print(total_tv)
```

TOTAL INVESTIDO EM R√ÅDIO:

```python
total_radio = tabela["Radio"].sum()
print(total_radio)
```

TOTAL INVESTIDO EM JORNAL:

```python
total_jornal = tabela["Jornal"].sum()
print(total_jornal)
```

TOTAL EM VENDAS:

```python
total_vendas = tabela["Vendas"].sum()
print(total_vendas)
```

```python
labels = ['TV', 'Radio', 'Jornal', 'Vendas']
totais = [total_tv, total_radio, total_jornal, total_vendas]

plt.pie(totais, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
```
